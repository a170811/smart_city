```python
# coding=utf-8

import numpy as np
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

from os import path
import pickle
from collections import defaultdict
```

```python
wu_feas = pickle.load(open('../logs/wu_selec_res.pkl', 'rb'))
chi_feas = pickle.load(open('../logs/chi_selec_res.pkl', 'rb'))
id2char = pickle.load(open('../logs/id2char.pkl', 'rb'))


print(id2char)
def count_feas(selec_feas, weight):
    counted_dict = defaultdict(int)
    weight = float(weight)
    
    for feas in selec_feas:
        for fea in feas:
            counted_dict[fea] += weight
    return counted_dict


def convert(target):
    res = {id2char[k]: v for k, v in target.items()}
    return res

wu_count = count_feas(wu_feas, 5)
chi_count = count_feas(chi_feas, 5)

# wu_count = convert(wu_count)
# chi_count = convert(chi_count)

print(wu_count)
print(chi_count)
```

```python
def plot_wordcloud(f_dict):
    wordcloud = WordCloud(width=800, height=400, max_font_size=80, max_words=40, background_color="white").generate_from_frequencies(f_dict)
    plt.figure(figsize=(18, 9))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
    
plot_wordcloud(wu_count)
plot_wordcloud(chi_count)
```
